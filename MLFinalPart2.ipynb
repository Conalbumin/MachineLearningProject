```python
#Cau 2
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

# Chọn đặc trưng và mục tiêu
X = data.drop('Class', axis=1)
y = data['Class']

# Chuyển đổi các đặc trưng kiểu chuỗi thành số bằng mã hóa one-hot
X_encoded = pd.get_dummies(X)

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Xây dựng và đánh giá các mô hình
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'Support Vector Machine': SVC(kernel='linear', C=1)
}

for model_name, model in models.items():
    # Huấn luyện mô hình
    model.fit(X_train, y_train)

    # Dự đoán trên tập kiểm tra
    y_pred = model.predict(X_test)

    # Đánh giá hiệu suất mô hình
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model_name} Accuracy: {accuracy:.2f}")
    print(f"{model_name} Classification Report:\n{classification_report(y_test, y_pred)}\n")

# Hiển thị biểu đồ so sánh hiệu suất
accuracy_values = [accuracy_score(y_test, model.predict(X_test)) for model in models.values()]
models_names = list(models.keys())

plt.figure(figsize=(10, 6))
plt.bar(models_names, accuracy_values, color=['blue', 'green', 'orange', 'red'])
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Comparison')
plt.show()
```

3)	Sử dụng Feed Forward Neural Network và Reccurent Neural Network (hoặc mô thuộc loại này) để giải quyết bài toán;

```python
# Feed Forward Neural Network (FNN):
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
from tensorflow import keras
from tensorflow.keras import layers

# Chọn đặc trưng và mục tiêu
X = data.drop('Class', axis=1)
y = data['Class']

# Chuyển đổi mục tiêu thành dạng số bằng Label Encoding
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Chuyển đổi các đặc trưng kiểu chuỗi thành số bằng mã hóa one-hot
X_encoded = pd.get_dummies(X)

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

# Xây dựng mô hình Feed Forward Neural Network
model_fnn = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(len(label_encoder.classes_), activation='softmax')
])

# Compile the model
model_fnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model_fnn.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model on the test set
y_pred_probs = model_fnn.predict(X_test)
y_pred_fnn = y_pred_probs.argmax(axis=1)  # Get the index of the highest probability for each sample

accuracy_fnn = accuracy_score(y_test, y_pred_fnn)
print(f"Feed Forward Neural Network Accuracy: {accuracy_fnn:.2f}")
print(f"Feed Forward Neural Network Classification Report:\n{classification_report(y_test, y_pred_fnn)}")
```

```python
# Recurrent Neural Network (RNN):
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Chọn đặc trưng và mục tiêu
X = data.drop('Class', axis=1)
y = data['Class']

# Chuyển đổi mục tiêu thành dạng số bằng Label Encoding
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Chuyển đổi các đặc trưng kiểu chuỗi thành dạng chuỗi văn bản (tokenizer)
tokenizer = Tokenizer()
X_text = X.apply(lambda row: ' '.join(row), axis=1)
tokenizer.fit_on_texts(X_text)
X_sequences = tokenizer.texts_to_sequences(X_text)
X_padded = pad_sequences(X_sequences)

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X_padded, y_encoded, test_size=0.2, random_state=42)

# Xây dựng mô hình Recurrent Neural Network
model_rnn = keras.Sequential([
    Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=64),
    SimpleRNN(64),
    Dense(len(label_encoder.classes_), activation='softmax')
])

# Compile the model
model_rnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model_rnn.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model on the test set
y_pred_probs_rnn = model_rnn.predict(X_test)
y_pred_rnn = y_pred_probs_rnn.argmax(axis=1)  # Get the index of the highest probability for each sample

accuracy_rnn = accuracy_score(y_test, y_pred_rnn)
print(f"Recurrent Neural Network Accuracy: {accuracy_rnn:.2f}")
print(f"Recurrent Neural Network Classification Report:\n{classification_report(y_test, y_pred_rnn)}")
```
